```python
import pandas as pd
import re
import requests
import nltk
from bs4 import BeautifulSoup

from wrangle import get_all_episode_urls
```

```python
series_url = "http://www.chakoteya.net/NextGen/episodes.htm"
```

```python
df = get_all_episode_urls(series_url)
```

```python
df.head()
```

```python
df["line"] = df.lines.str.replace(r'^[A-Z]+:', "", regex=True)
df.head()
```

```python
df.lines = df.lines.str.replace(r":.+", "", regex=True)
```

```python
character = df.lines.str.replace(r"\s.*", "", regex=True)
```

```python
df["character"] = character
```

```python
df = df[["episode_name", "line", "character"]]
df
```

```python
df.to_csv("tng.csv")
```
